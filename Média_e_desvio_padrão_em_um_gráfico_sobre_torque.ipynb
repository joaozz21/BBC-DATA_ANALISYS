Para este item você deverá usar o arquivo [com dados de força e torque medidos no chão enquanto uma pessoa tenta ficar em pé durante um minuto](https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/balance.csv).

- Faça o gráfico do torque na direção y (My [Nm]) em função do tempo (Time [s]).

- Mostre na tela a média e o desvio-padrão do torque na direção y.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv('https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/balance.csv')

x = ds['Time[s]']
y = ds['My[Nm]']
media = y.mean()
desvio = y.std()

plt.figure()
plt.plot(x, y)
plt.axhline(media, color='r', linestyle='-', marker = '', label='Média')
plt.axhline(media + desvio, color='g', linestyle='--', label='Desvio padrão')
plt.axhline(media - desvio, color='g', linestyle='--')
plt.legend()
plt.xlabel('Tempo[s]')
plt.ylabel('Torque em Y[Nm]')
plt.title('Torque na direção y em função do tempo')
plt.show()
```
