**1)** Faça uma função que receba um vetor **t**,um valor **A**, um valor **b** e um valor **c** e retorne um outro vetor **x**.

$$x(t)  = A\sqrt{1+bt^2} + c$$

Além disso a função deve plotar o gráfico de **x** em função de **t**.

Teste para 
- *a)* t indo de 0 a 10 com 1000 pontos, A = 2, b = 0.5, c = 0; 
- *b)* t indo de 0 a 15 com intervalo de 0.2 entre os valores do vetor, A = 10, b = 0.2, c = 1
- *c*) t indo de -0,5 a 0,5 com 500 pontos, A = -3, b = -1.5, c = -10; 

```python
import numpy as np
import matplotlib.pyplot as plt

def fun(t,A,b,c):
  '''
  Retorna um vetor x e plota um gráfico de x em função de t

  Parâmetros
  ---------------------------------------------------------
  t:array
  vetor utilizado na função

  A:int
  Número inteiro utilizado na função

  b:float
  Número racional utilizado na funçãp

  c:int
  Número inteiro utilizado na função

  Retorna
  ---------------------------------------------------------
  x:array
  vetor criado a partir de uma função
  '''
  x= A*(1+b*t**2)**0.5+c

  plt.plot(t,x)
  plt.grid()
  plt.show()

  return x

#a)
t1=np.linspace(0,10,1000)  
A1=2
b1=0.5
c1=0
print(fun(t1,A1,b1,c1))

#b)
t2=np.arange(0,15,0.2)  
A2=10
b2=0.2
c2=1
print(fun(t2,A2,b2,c2))

#c)
t3=np.linspace(-0.5,0.5,500)  
A3=-3
b3=-1.5
c3=-10
print(fun(t3,A3,b3,c3))
```
