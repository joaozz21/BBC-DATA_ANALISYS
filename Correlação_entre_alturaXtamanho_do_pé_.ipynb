Faça o gráfico de dispersão do tamanho do pé (FootLen) em função da altura da pessoa (Height). Calcule (e mostre na tela) a correlação e plote a reta de regressão entre esses dados.

Dica: esse arquivo separa as suas colunas com uma tabulação. Para ler o arquivo a opção 'sep' deve ser '\t'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds= pd.read_csv('https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/BDSinfo.csv', sep='\t')

ds = ds.dropna(subset=['Height', 'FootLen'])

correlacao = ds['Height'].corr(ds['FootLen'])
m,b = np.polyfit(ds['Height'], ds['FootLen'], deg= 1)

print(f'Correlação entre altura e tamanho do pé: {round(correlacao*100,2)}%\n')

plt.figure()
plt.scatter(ds['Height'], ds['FootLen'])
plt.plot(ds['Height'], ds['Height']*m+b, color= 'red', label= 'Regressão')
plt.legend()
plt.title('Altura x Tamanho do pé')
plt.xlabel('Altura')
plt.ylabel('Tamanho do pé')
plt.show()
```
