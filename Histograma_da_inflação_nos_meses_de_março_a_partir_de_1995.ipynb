Para este item você deverá usar o arquivo com a [taxa de inflação mensal (IGP-DI) no Brasil de fevereiro de 1944 a junho de 2019](https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/inflacaoMensal.csv) (dados disponibilizados por Ipeadata).

- Calcule a mediana da inflação mensal nos meses de março a partir de 1995.

- Calcule a média da inflação mensal nos meses de março a partir de 1995.

- Calcule o desvio-padrão da inflação mensal nos meses de março a partir de 1995.

- Faça o histograma da inflação mensal nos meses de março a partir de 1995.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv('https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/inflacaoMensal.csv')

marcoDepoisDe95 = ds.query('Mês == 3 and Ano >= 1995')
media = marcoDepoisDe95['Inflação'].mean()
mediana = np.median(np.sort(marcoDepoisDe95['Inflação']))
desvioPadrao = marcoDepoisDe95['Inflação'].std()

print('A média é', round(media,2))
print('A mediana é', mediana)
print('O desvio padrão é', round(desvioPadrao,2))

plt.figure()
marcoDepoisDe95['Inflação'].hist()
plt.xlabel('Inflação')
plt.ylabel('Frequência')
plt.title('Histograma da Inflação Mensal nos meses de Março a partir de 1995')
plt.show()
```
