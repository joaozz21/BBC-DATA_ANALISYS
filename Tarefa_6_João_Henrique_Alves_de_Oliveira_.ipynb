```python

# Ex1
import numpy as np
import matplotlib.pyplot as plt

#Definindo o intervalo da função e a fórmula

t=np.arange(0,10.05,0.5)
v=np.exp(-0.5*t)*np.cos(2*np.pi*0.8*t)

#Gerando o gráfico 

plt.figure()
plt.plot(t, v)
plt.xlabel('tempo(s) ') 
plt.ylabel('tensão(v) ') 
plt.title('Gráfico da tensão em função do tempo') 
plt.show()
```

```python

#Ex2
import numpy as np
import matplotlib.pyplot as plt

#Definindo o intervalo da função e a fórmula

C=np.linspace(-20,100,100)
T=9/5*C+32

#Gerando o gráfico 

plt.figure()
plt.plot(C, T)
plt.xlabel('Celsius(C) ') 
plt.ylabel('Fahrenheit(F) ') 
plt.title('Gráfico da temperatura em Fahrenheit em função de Celsius') 
plt.show()
```

```python
#Ex3
import numpy as np
import matplotlib.pyplot as plt

#Definindo o intervalo da função e a fórmula

y=np.arange(1994,2022) 
p=800000*np.exp((y-1994)/40)

#Gerando o gráfico 

plt.figure()
plt.plot(y,p) 
plt.xlabel('anos(y)') 
plt.ylabel('população(p)') 
plt.show()
```
