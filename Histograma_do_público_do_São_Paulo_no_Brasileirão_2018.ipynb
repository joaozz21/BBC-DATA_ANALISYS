- Escolha um time e faça o histograma da distribuição do público nos jogos em que este time foi o mandante.

- Calcule qual foi o público médio nos jogos em que o time escolhido foi o mandante.

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv("https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/tabelaBrasileirao2018.csv")
publicoSaoPaulo = ds.query("Mandante == 'São Paulo'")['Público']

plt.figure()
publicoSaoPaulo.hist()
plt.title("Histograma da distribuição do público nos jogos em que o time São Paulo foi o mandante")
plt.xlabel("Público")
plt.ylabel("Frequência")
plt.show()

print("O público médio nos jogos em que o time São Paulo foi o mandante foi de", round(publicoSaoPaulo.mean()), "pessoas")
```
