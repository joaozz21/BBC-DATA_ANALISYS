 A acidez $A(x)$ de uma solução de hidróxido de magnésio em ácido clorídrico, sob certas condições experimentais, é dada pela equação:

$$A(x) = x^3 + 3x^2 − 54$$
na qual $x$ é a concentração de íons hidrônio.

Pede-se:

1) Gere o gráfico de $A(x)$ em função de $x$ para 0 $\leqslant x \leqslant 8$.

2) Determine a concentração $x$ do íon de hidrônio que resulta em solução saturada (i.e., com acidez nula). Acrescente uma instrução que gere um ponto vermelho no gráfico correspondente à acidez nula da solução.

```python
import numpy as np
import matplotlib.pyplot as plt

#Definindo o intervalo da função e a fórmula

x = np.arange(0,8.1,0.1)
y = x**3+3*(x**2)-54

#Gerando o gráfico

plt.figure()
plt.plot(x,y)
plt.plot(x[y==0],0,'ro',label='Acidez Nula')
plt.xlabel('Concentração de íons hidrônio ')
plt.ylabel('Acidez')
plt.title('Acidez de uma solução de hidróxido de magnésio em ácido clorídrico')
plt.legend()
plt.grid()
plt.show()
```

 Faça o gráfico de:
    $$f(x)=|x−2|+|2x+1|−x−6$$
    
 Faça o gráfico com $x$ variando de -10 a 10.

 Mostre o menor valor de $x$ tal que $f(x)>0$ e $x>0$.

```python
#Definindo o intervalo da função e a fórmula

x = np.arange(-10,10.5,0.5)
y = np.abs(x-2) + np.abs(2*x+1) - x - 6

xplus = x[x>0]
yplus = np.abs(xplus-2) + np.abs(2*xplus+1) - xplus - 6
menor_valor = np.min(xplus[yplus>0])
print(np.min(y[y>0]))

#Gerando o gráfico
plt.figure()
plt.plot(x,y)
plt.plot(menor_valor,np.min(y[y>0]),'ro',label='Menor valor de x quando x>0<f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
