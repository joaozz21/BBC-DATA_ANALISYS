 Faça o gráfico da Renda [dos jogos do campeonato brasileiro de 2018](https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/tabelaBrasileirao2018.csv) em função do público de cada jogo. Calcule a correlação entre essas grandezas.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv('https://raw.githubusercontent.com/joaozz21/BasesComputacionais/refs/heads/master/dados/tabelaBrasileirao2018.csv')

m,b = np.polyfit(ds['Público'], ds['Renda (R$)'], deg=1)

correlacao = ds['Público'].corr(ds['Renda (R$)'])
print(f'Correlação entre público e renda: {round(correlacao*100,1)}%\n')

plt.figure()
plt.scatter(ds['Público'], ds['Renda (R$)'])
plt.plot(ds['Público'], ds['Público'] * m + b, color= 'red', label= 'Regrassão Linear')
plt.legend()
plt.title('Renda x Público')
plt.xlabel('Público')
plt.ylabel('Renda (R$)')
plt.show()
```
