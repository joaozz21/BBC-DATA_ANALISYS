```python

import numpy as np
import matplotlib.pyplot as plt
#a) 
#Definindo o intervalo da função e a fórmula

t= np.arange(0,5,0.1)
h= - t**2+4*t

#Gerando o gráfico 

plt.figure()
plt.plot(t, h)
plt.xlabel('tempo(s) ') 
plt.ylabel('altura(m)') 
plt.title('Gráfico da altura em função do tempo') 
plt.show()

#b) 
#Executando a função que busca o valor máximo de h
hmax=np.max(h) 
print(hmax)
```
